```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException,ElementNotInteractableException
import requests
import re
from datetime import datetime
import warnings
warnings.filterwarnings("ignore")
```

```python
# opening the url with selenium

driver = webdriver.Chrome(r"C:\Desktop\DT\chromedriver.exe")

# opening yatra.com page
url = "https://www.yatra.com/"
driver.get(url)

time.sleep(2)
```

```python
df=pd.DataFrame()
date=[]
city_list=[' delhi',' mumbai',' bangalore',' pune',' chennai',' Hyderabad',' Ahmedabad',' Cochin',' Goa']
for i in city_list:
    city_list.remove(i)
    for j in city_list:
        
        driver.find_element_by_xpath('//li[@class="datepicker flex1"]').click()
        time.sleep(1)
        date1=driver.find_elements_by_xpath('//tbody[@class="BE_flight_origin_date"]//tr//td')
        for k in date1:
            if k.get_attribute('data-date') !=None and k.get_attribute('data-date') !="":
                date.append(k.get_attribute('data-date'))
        for k in date:
        
                origin=driver.find_element_by_xpath('//input[@class="required_field cityPadRight ac_input origin_ac"]')
                for car in i:
                    time.sleep(0.2)
                    origin.send_keys(car)
                origin.send_keys(Keys.ENTER)

                time.sleep(0.5)
                try:
                    dest=driver.find_element_by_xpath('//input[@class="required_field cityPadLeft ac_input dest_ac"]')
                except NoSuchElementException:
                    time.sleep(1)
                    dest=driver.find_element_by_xpath('//input[@class="required_field cityPadLeft ac_input dest_ac"]')
                for car in j:
                    time.sleep(0.2)
                    dest.send_keys(car)
                dest.send_keys(Keys.ENTER)
                time.sleep(0.5)
                try:
                    driver.find_element_by_xpath('//li[@class="datepicker flex1"]').click()
                except:
                    print('ok')
                time.sleep(0.5)
                driver.find_element_by_xpath(f'//td[@data-date="{k}"]').click()
                driver.find_element_by_xpath('//input[@class="js_ripple search-btn "]').click() #clicking search buttom 
                time.sleep(0.5)
                alpha=0
                while alpha<5:
                    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                    time.sleep(0.4)
                    alpha+=1
                stops=[i.text for i in driver.find_elements_by_xpath('//div[@class=" font-lightgrey fs-10 tipsy i-b fs-10"]')]
                airline=[i.text for i in driver.find_elements_by_xpath('//span[@class="i-b text ellipsis"]')]                    
                city = [i.text for i in driver.find_elements_by_xpath('//p[@class="fs-10 font-lightgrey no-wrap city ellipsis"]')]
                d_t=[i.text for i in driver.find_elements_by_xpath('//div[@class="i-b pr"]')]
                a_t=[i.text for i in driver.find_elements_by_xpath('//p[@class="bold fs-15 mb-2 pr time"]')]
                t_t=[i.text for i in driver.find_elements_by_xpath('//p[@class="fs-12 bold du mb-2"]')]                    
                price=[i.text for i in driver.find_elements_by_xpath('//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')]
                stops=[i.text for i in driver.find_elements_by_xpath('//div[@class=" font-lightgrey fs-10 tipsy i-b fs-10"]')]
                date1=[k for i in range(len(price))]
                a=1
                destination=[]
                source=[]
                for l in city:
                    if a%2==0:
                        destination.append(l)
                    else:
                        source.append(l)
                    a+=1
                temp=pd.DataFrame()
                temp['airline']=airline
                temp['date']=date1
                temp['source']=source
                temp['destination']=destination
                temp['price']=price
                temp['d_t']=d_t
                temp['a_t']=a_t
                temp['t_t']=t_t
                temp['stops']=stops
                temp['price']=price
                
                df=pd.concat([df,temp],axis=0)
                url = "https://www.yatra.com/"
                driver.get(url)

                time.sleep(2)
```

```python
def ind(l,a):
    for b in range(len(l)):
        if l[b] == a:
            break
    return b
```
